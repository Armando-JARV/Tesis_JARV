# Clasificiación consenso

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame()
df_PTP1B = pd.read_csv('STRAT_PTP1B_3Metric_WSMILES_REVIEWD.csv') 
df_AR = pd.read_csv('STRAT_AR_3Metric_WSMILES_REVIEWD.csv')
df_PTP1B
```

```python
print (len (df_PTP1B))
df_PTP1B = df_PTP1B[(df_PTP1B['LELP_MOE'] >= 0) & (df_PTP1B['LELP_Ledock'] >= 0)  & (df_PTP1B['LELP_VINA'] >= 0)]
df_PTP1B = df_PTP1B.reset_index(drop=True)
print (len (df_PTP1B))

#df_AR = df_AR[(df_AR['LELP_MOE'] >= -0) & (df_AR['LELP_Ledock'] >= -0)  & (df_AR['LELP_VINA'] >= 0)]
print (len (df_AR))
df_AR = df_AR[(df_AR['LELP_MOE'] >= -0) & (df_AR['LELP_Ledock'] >= -0)  & (df_AR['LELP_VINA'] >= 0)]
df_AR = df_AR.reset_index(drop=True)
print (len (df_AR))
```

```python
# Fusionar df_Ledock
df_Ledock = df_PTP1B[['DB_ID', 'LELP_Ledock']].rename(columns={'LELP_Ledock': 'LELP_Ledock_PTP1B'}).merge(
    df_AR[['DB_ID', 'LELP_Ledock']].rename(columns={'LELP_Ledock': 'LELP_Ledock_AR'}),
    on='DB_ID',
    how='inner'
)

# Fusionardf_VINA
df_VINA = df_PTP1B[['DB_ID', 'LELP_VINA']].rename(columns={'LELP_VINA': 'LELP_VINA_PTP1B'}).merge(
    df_AR[['DB_ID', 'LELP_VINA']].rename(columns={'LELP_VINA': 'LELP_VINA_AR'}),
    on='DB_ID',
    how='inner' 
)

print(df_Ledock)
print(df_VINA)
```

```python
df_MOE = df_PTP1B[['DB_ID', 'LELP_MOE']].rename(columns={'LELP_MOE': 'LELP_MOE_PTP1B'}).merge(
    df_AR[['DB_ID', 'LELP_MOE']].rename(columns={'LELP_MOE': 'LELP_MOE_AR'}),
    on='DB_ID',
    how='inner'  # 'outer', 'left' o 'right' 
)

print(df_MOE)
```

```python
# Aqui es importante el
#ascending=False/True
#False es para rankearlos de mayores a menores (descendente)
```

```python
df_MOE['R_MOE_PTP1B'] = df_MOE['LELP_MOE_PTP1B'].rank(ascending=True, method='dense')
df_MOE['R_MOE_AR'] = df_MOE['LELP_MOE_AR'].rank(ascending=True, method='dense')
# df_MOE[['DB_ID', 'S_MOE_PTP1B', 'R_MOE_PTP1B']]
# df_MOE[['DB_ID', 'S_MOE_AR', 'R_MOE_AR']]
```

```python
# Aplicar ranking a df_Ledock
df_Ledock['R_Ledock_PTP1B'] = df_Ledock['LELP_Ledock_PTP1B'].rank(ascending=True, method='dense')
df_Ledock['R_Ledock_AR'] = df_Ledock['LELP_Ledock_AR'].rank(ascending=True, method='dense')

# Aplicar ranking a df_VINA
df_VINA['R_VINA_PTP1B'] = df_VINA['LELP_VINA_PTP1B'].rank(ascending=True, method='dense')
df_VINA['R_VINA_AR'] = df_VINA['LELP_VINA_AR'].rank(ascending=True, method='dense')

# Visualizar los resultados
print(df_Ledock[['DB_ID', 'LELP_Ledock_PTP1B', 'R_Ledock_PTP1B']])
print(df_Ledock[['DB_ID', 'LELP_Ledock_AR', 'R_Ledock_AR']])

print(df_VINA[['DB_ID', 'LELP_VINA_PTP1B', 'R_VINA_PTP1B']])
print(df_VINA[['DB_ID', 'LELP_VINA_AR', 'R_VINA_AR']])
```

```python
tau = 10
df = df_VINA[['DB_ID']].copy()

df['ECR_PTP1B'] = (1 / tau) * (
    np.exp(-df_MOE['R_MOE_PTP1B'] / tau) +
    np.exp(-df_VINA['R_VINA_PTP1B'] / tau) +
    np.exp(-df_Ledock['R_Ledock_PTP1B'] / tau))

df["R_ECR_PTP1B"] = df["ECR_PTP1B"].rank(ascending=False, method="dense")

print(df[["DB_ID", "ECR_PTP1B", "R_ECR_PTP1B"]])
```

```python
df
```

```python
tau = 10

df['ECR_AR'] = (1 / tau) * (
    np.exp(-df_MOE['R_MOE_AR'] / tau) +
    np.exp(-df_VINA['R_VINA_AR'] / tau) +
    np.exp(-df_Ledock['R_Ledock_AR'] / tau)
)

df["R_ECR_AR"] = df["ECR_AR"].rank(ascending=False, method="dense")

print(df[["DB_ID", "ECR_AR", "R_ECR_AR"]])
```

```python
df_cons = df[['DB_ID', "ECR_PTP1B", "R_ECR_PTP1B", "ECR_AR", "R_ECR_AR"]].copy()
df_cons
```

```python
# Paso 1: Promediar los rankings de cada objetivo
df_cons["Average_rank"] = df_cons[["R_ECR_PTP1B", "R_ECR_AR"]].mean(axis=1)

# Paso 2: Ranking final (como en Excel con SUMPRODUCT / COUNTIF)
df_cons["Final_consensus_rank"] = df_cons["Average_rank"].rank(ascending=True, method="dense")

# Ver resultados ordenados
final_result = df_cons.sort_values("Final_consensus_rank")[[
    "DB_ID", "R_ECR_PTP1B", "R_ECR_AR", "Average_rank", "Final_consensus_rank"
]]

final_result
```

```python
final_result.to_csv('Resultado_consenso_LELP_ECR_false_ok.csv', index=False)
```
